```python
def avg_toi(df):
    df['Avg TOI'] = round(df['TOI']/df['GP'], 3)
    return df

def points_per_60(df):
    df['Pts per 60'] = round(df['PTS']/df['Avg TOI'], 3)
    return df

def points_per_game(df):
    df['Pts per Game'] = round(df['PTS']/df['GP'], 3)
    return df

def pp60(df):
    df['P/60'] = round(df["PTS"] * 60 / df['Avg TOI'], 3)
    return df

def p_per60(df): # THIS ONE 
    df['PP60'] = round(df["PTS"] * 60 / df['TOI'], 3)
    return df
```
